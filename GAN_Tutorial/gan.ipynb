This tutorial will explain how a Generative Adversarial Network (GAN) works.

<img src="gan.png" alt="Drawing" style="width: 400px;"/>

```python
from gan import *
```

First, let us create a generator where fake images are generated.

```python
fake_images = generator()
```

Now, we create another part of the network which sample real images from data.

```python
real_images = real()
```

Then, a discriminator take the images (both fake and real) and output a probability of whether they are real or not.

```python
p_fake_is_real = discriminator(fake_images)
```

```python
p_real_is_real = discriminator(real_images)
```

For training, we want the network to predict correctly which are real images and which are not, by maximising the probability of real images are real and mininising the probability of fake images are real. 

```python
discriminator_objective = maximise(p_real_is_real) + minimise(p_fake_is_real)
```

at the same time, we want the generator generate "good" fake images that can cheat the disciminator, by maximising the probability that fake images are real.

```python
generator_objective = maximise(p_fake_is_real)
```

Let's train the network and observe the (fake) images generated on process.

```python
%matplotlib inline
train(generator_objective,discriminator_objective)
```
